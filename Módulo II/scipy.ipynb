O **módulo `stats`** dentro do `SciPy` é responsável por fornecer ferramentas estatísticas essenciais para análise de dados, desde estatísticas descritivas até testes estatísticos avançados.


**Principais Funcionalidades do `scipy.stats`**  
O módulo `stats` oferece uma vasta gama de funções estatísticas. Algumas das mais utilizadas incluem:

- **Testes Estatísticos:** Teste $t$, $\chi^{2}$, Shapiro-Wilk  
- **Correlação e Regressão:** Pearson, Spearman, regressão linear

## (5.1) Distribuições de Probabilidade

### (5.1.1) Geração de Variáveis Pseudo-aleatórias

- **Distribuição Normal:** $$
f(x) = \frac{1}{\sigma\sqrt{2\pi}} \exp\left\{ -\dfrac{1}{2}\left(\dfrac{x-\mu}{\sigma}\right)^2 \right\}$$

```python
# Geração
n, mu, sigma = 1000, 0, 1
x = np.array([stats.norm.rvs(loc=mu, scale=sigma) for _ in range(n)]).reshape(-1, 1)
```

```python
# Configurações da Figura
fig, axes = plt.subplots(2, 2, figsize=(16, 8), dpi=600)

# Alterando o formato do vetor de eixos
axes = axes.flatten()

# Plot
axes[0].plot(x)
axes[0].set_xlabel("Index", fontsize=12)
axes[0].set_ylabel("X", fontsize=12)
axes[0].spines["top"].set_visible(False)
axes[0].spines["right"].set_visible(False)

# Histograma
axes[1].hist(x, bins=30)
axes[1].set_xlabel("X", fontsize=12)
axes[1].set_ylabel("Frequência", fontsize=12)
axes[1].spines["top"].set_visible(False)
axes[1].spines["right"].set_visible(False)

# QQ Plot
stats.probplot(x.flatten(), dist="norm", plot=axes[2])
axes[2].set_xlabel("Quantil Teórico", fontsize=12)
axes[2].set_ylabel("Quantil Amostral", fontsize=12)
axes[2].spines["top"].set_visible(False)
axes[2].spines["right"].set_visible(False)

# Boxplot
axes[3].boxplot(x)
axes[3].set_xticklabels([""])
axes[3].set_ylabel("X", fontsize=12)
axes[3].spines["top"].set_visible(False)
axes[3].spines["right"].set_visible(False)

# Exibindo o gráfico e ajuste do layout
fig.tight_layout()
plt.show()
```

```python
# Função Densidade de Probilidade & Função de Distribuição
pdf = stats.norm.pdf(np.sort(x, axis=0), loc=mu, scale=sigma)
cdf = stats.norm.cdf(np.sort(x, axis=0), loc=mu, scale=sigma)

# Configurações da Figura
fig, axes = plt.subplots(1, 2, figsize=(12, 4), dpi=600)

# Função Densidade de Probilidade
axes[0].plot(np.sort(x, axis=0), pdf)
axes[0].set_xlabel("X", fontsize=12)
axes[0].set_ylabel("Densidade de Probabilidade", fontsize=12)
axes[0].spines["top"].set_visible(False)
axes[0].spines["right"].set_visible(False)

# Função de Distribuição
axes[1].plot(np.sort(x, axis=0), cdf)
axes[1].set_xlabel("X", fontsize=12)
axes[1].set_ylabel("Função de Distribuição", fontsize=12)
axes[1].spines["top"].set_visible(False)
axes[1].spines["right"].set_visible(False)

# Exibição do gráfico
plt.show()
```
